6. Mini Bio Task : DNA Sequence analyzer

```python
sequence = "ATGCGATACTTGGAATTGG"

a = sequence.count("A")
t = sequence.count("T")
g = sequence.count("G")
c = sequence.count("C")

print("A:", a, "T:", t, "G:", g, "C:", c)

gc_content = ((g + c) / len(sequence)) * 100
print("GC Content:", round(gc_content, 2), "%")

if gc_content > 50:
    print("High GC gene â€” stable structure")
else:
    print("Low GC gene â€” AT-rich region")
```
